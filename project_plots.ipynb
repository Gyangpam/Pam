```python
import pandas as pd
data=pd.read_excel('withdfig.xlsx')
t=list(data.Time)
a1=list(data.A1)
a2=list(data.A2)
a3=list(data.A3)
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rc('xtick', labelsize=25)
matplotlib.rc('ytick', labelsize=25)
plt.figure(figsize=(20,12))
plt.plot(t, a1, 'r*', linewidth=1)
plt.plot(t, a2, 'b+', linewidth=1)
plt.plot(t, a3, 'g-', linewidth=1)
plt.minorticks_on()
plt.grid(True, which="both")
plt.axis([0,30,-10,100])
plt.legend(['G1: Rotor Angel with Reference to Reference machine angle in deg', 
            'G2: Rotor Angel with Reference to Reference machine angle in deg', 
            'G3: Rotor Angel with Reference to Reference machine angle in deg'], fontsize="xx-large")
plt.xlabel('Time [s]', size=25)
plt.ylabel('Rotor Angle [deg]', size=25)

```
